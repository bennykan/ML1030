```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import glob
import os
import re
```

```python
path = r'C:\Users\benny\OneDrive\Machine Learning\ML1030\Final Project\Final Project\Data' # use your path
all_csv_files = glob.glob(os.path.join(path, "*.csv"))
all_xlsx_files = glob.glob(os.path.join(path, "*.xlsx"))

#(head, tail) = os.path.split(all_xlsx_files[0])
#print(tail)
#name = re.search('(.+?).xlsx', tail).group(1)
#print(name)
#data_xls = pd.read_excel(all_xlsx_files[0])
#data_xls.to_csv(name+'.csv', encoding='utf-8')

#for filename in all_xlsx_files[0:1]:
#    data_xls = pd.read_excel(filename, 'Sheet2', index_col=None)
#    data_xls.to_csv('csvfile.csv', encoding='utf-8')

print(len(all_csv_files))
print(len(all_xlsx_files))
```

```python
li_csv = pd.DataFrame()
i = 1

for filename in all_csv_files:
    (head, tail) = os.path.split(filename)
    teams = re.search('NFL-(.+?)-data', tail).group(1)
    #print(tail)
    #print(teams)
    away=teams.split('@')[0]
    home=teams.split('@')[1]
    #print(away)
    #print(home)
    csv_df = pd.read_csv(filename)
    csv_df['Away'] = away
    csv_df['Home'] = home
    csv_df['Game'] = teams
    csv_df['Game Number'] = i
    i = i+1
    li_csv = pd.concat([li_csv,csv_df],sort=None)

print(li_csv.shape)    
    
li_xlsx = pd.DataFrame()

for filename in all_xlsx_files:
    (head, tail) = os.path.split(filename)
    teams = re.search('NFL-(.+?)-data', tail).group(1)
    #print(tail)
    #print(teams)
    away=teams.split('@')[0]
    home=teams.split('@')[1]
    #print(away)
    #print(home)
    xlsx_df = pd.read_excel(filename)
    xlsx_df['Away'] = away
    xlsx_df['Home'] = home
    xlsx_df['Game'] = teams
    xlsx_df['Game Number'] = i
    i = i+1
    li_xlsx = pd.concat([li_xlsx,xlsx_df],sort=None)

print(li_xlsx.shape)

all_data = pd.concat([li_csv,li_xlsx], sort=None, ignore_index=True)

print(all_data)
all_data.to_csv(path+"\\all_data.csv")
```
