```python
def calculate_juice(away,home):
  if(away < 0):
    decimal_odds_away = 1+100/abs(away)
  else:
    decimal_odds_away = 1+abs(away)/100

  if(home < 0):
    decimal_odds_home = 1+100/abs(home)
  else:
    decimal_odds_home = 1+abs(home)/100
    
  return (1-(decimal_odds_away*decimal_odds_home)/(decimal_odds_away+decimal_odds_home))*100
```
