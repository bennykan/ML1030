```python
import pandas as pd
import numpy as np
from datetime import datetime
from pytz import timezone
import pytz
import glob
import os
import re
```

```python
nfl_games_file = r'C:\Users\benny\OneDrive\Machine Learning\ML1030\Final Project\Final Project\sportsbook_data_ALL.csv'
nfl_games_df = pd.read_csv(nfl_games_file)
```

```python
nfl_games_df.head()
```

```python
def calculate_juice(away,home):
  if(away < 0):
    decimal_odds_away = 1+100/abs(away)
  else:
    decimal_odds_away = 1+abs(away)/100

  if(home < 0):
    decimal_odds_home = 1+100/abs(home)
  else:
    decimal_odds_home = 1+abs(home)/100
    
  return (1-(decimal_odds_away*decimal_odds_home)/(decimal_odds_away+decimal_odds_home))*100
```

```python
nfl_games_df['Juice Moneyline']=""
nfl_games_df['Juice Pointspread']=""
nfl_games_df['Juice Total']=""

for index, row in nfl_games_df.iterrows():
    try:
        nfl_games_df.loc[index,'Juice Moneyline']=calculate_juice(float(row['Away Moneyline']),float(row['Home Moneyline']))
        nfl_games_df.loc[index,'Juice Pointspread']=calculate_juice(float(row['Away Pointspread odds']),float(row['Home Pointspread odds']))
        nfl_games_df.loc[index,'Juice Total']=calculate_juice(float(row['Over odds']),float(row['Under odds']))
    except Exception:
        pass
```

```python
nfl_games_juice_df['Book'] = nfl_games_juice_df['Book'].map(lambda x: x.lower())
nfl_games_juice_df.Book.replace('juilettbook', 'juliettbook', inplace=True) 

nfl_games_juice_pregame_df = nfl_games_juice_df[nfl_games_juice_df['Pre-Game']==True]
nfl_games_juice_livegame_df = nfl_games_juice_df[nfl_games_juice_df['Pre-Game']==False]

grouped_pregame = nfl_games_juice_pregame_df.groupby('Book')
moneyline_pregame = grouped_pregame['Juice Moneyline'].agg([np.max]).sort_values(by=['amax'],ascending=False)
pointspread_pregame = grouped_pregame['Juice Pointspread'].agg([np.max]).sort_values(by=['amax'],ascending=False)
total_pregame = grouped_pregame['Juice Total'].agg([np.max]).sort_values(by=['amax'],ascending=False)
print(moneyline_pregame)
print(pointspread_pregame)
print(total_pregame)

grouped_livegame = nfl_games_juice_livegame_df.groupby('Book')
moneyline_livegame = grouped_livegame['Juice Moneyline'].agg([np.max]).sort_values(by=['amax'],ascending=False)
pointspread_livegame = grouped_livegame['Juice Pointspread'].agg([np.max]).sort_values(by=['amax'],ascending=False)
total_livegame = grouped_livegame['Juice Total'].agg([np.max]).sort_values(by=['amax'],ascending=False)
print(moneyline_livegame)
print(pointspread_livegame)
print(total_livegame)
```
