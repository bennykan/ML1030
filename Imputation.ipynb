```python
import pandas as pd
import glob
import os
home_pointspread_data = pd.read_csv (r'C:\Users\shabe\home_pointspread_time_series_df.csv')
away_pointspread_data = pd.read_csv (r'C:\Users\shabe\away_pointspread_time_series_df.csv')
under_odds_data = pd.read_csv (r'C:\Users\shabe\under_odds_time_series_df.csv')
over_odds_data = pd.read_csv (r'C:\Users\shabe\over_odds_time_series_df.csv')
home_moneyline_data = pd.read_csv (r'C:\Users\shabe\home_moneyline_time_series_df.csv')
away_moneyline_data = pd.read_csv (r'C:\Users\shabe\away_moneyline_time_series_df.csv')
home_pointspread_data.head()
```

```python
pointspread_data = home_pointspread_data.append(home_pointspread_data, ignore_index=True)
total_points_data = over_odds_data.append(under_odds_data, ignore_index=True)
moneyline_data = home_moneyline_data.append(away_moneyline_data, ignore_index=True)
print(len(pointspread_data.index))
print(len(total_points_data.index))
print(len(moneyline_data.index))
pointspread_data.dropna(thresh=10,inplace=True)
total_points_data.dropna(thresh=10,inplace=True)
moneyline_data.dropna(thresh=10,inplace=True)
print(len(moneyline_data.index))
```

```python
#impute missing values 
pointspread_data.fillna(method='ffill', inplace=True)
total_points_data.fillna(method='ffill', inplace=True)
moneyline_data.fillna(method='bfill', inplace=True)
```

```python
from pandas import ExcelWriter
print(len(pointspread_data.index))
print(len(total_points_data.index))
writer = ExcelWriter('time_series_impute.xlsx')
pointspread_data.to_excel(writer,'pointspread')
total_points_data.to_excel(writer,'total_points')
writer.save()
```
